## **RELATIONSHIP ANALYSIS**

```
#CORRELATION MATRIX
corelation=Altered_data.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True,cmap="coolwarm")
```

**blue is negative corelation and red is positive corelation.**

```
sns.pairplot(Altered_data)
```

```
#scatter plot
sns.relplot(x = 'Sales',y = 'Profit',hue='Region',data=Altered_data,color='black')
sns.relplot(x = 'Sales',y = 'Profit',hue='State',data=Altered_data,color='red')
```

## **CONCLUSION**


**=>Strong negative corelation between discount and profit.
